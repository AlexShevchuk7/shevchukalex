```python
# Pytorch datasets
def generate_shapes(X=64, Y=64):
    img = np.ones((X, Y, 3))
    shape = np.random.choice(['square', 'circle'])
    k = np.random.choice([2, 4, 8])
    intens = np.random.choice([0.1, 0.25, 0.5])
    channel = np.random.choice([0, 1, 2])

    for i in range(X):
        for j in range(Y):
            if shape == 'square':
                if abs(i - int(X / 2)) <= X / (2 * k) and abs(j - int(Y / 2)) <= Y / (2 * k):
                    img[i, j, channel] = intens
                    
            if shape == 'circle':
                if (i - int(X / 2)) ** 2 + (j - int(Y / 2)) ** 2 <= X ** 2 / (k * k):
                    img[i, j, channel] = intens
                
    return img
```

```python
class ShapesDataset(Dataset):
    def __init__(self, generator, num_samples):
        self.data = torch.stack([torch.Tensor(generator()) for k in range(num_samples)])
        self.shape = self.__getshape__()
        self.size = self.__getsize__()
 
    def __getitem__(self, index):
        return self.data[index]
 
    def __len__(self):
        return len(self.data)
    
    def __getshape__(self):
        return (self.__len__(), *self.__getitem__(0)[0].shape)
    
    def __getsize__(self):
        return (self.__len__())
```

```python
def test(net):
    net.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, labels in test_loader:
            labels = labels.to(device)
            output = net(data.to(device))
            test_loss += criterion(output, labels.to(device)).item()
            preds = output.data.max(1, keepdim=True)[1]
            correct += preds.eq(labels.data.view_as(preds.to(device))).sum()
        test_loss /= len(test_loader.dataset)
        test_losses.append(test_loss)
        
    print('\nTest set: Avg. loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset), 
        100. * correct / len(test_loader.dataset)))
```

```python
class ConvEncoder(nn.Module):
    def __init__(self, num_classes):
        super(ConvEncoder, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1 = nn.Linear(320, 50)
        self.fc2 = nn.Linear(50, num_classes)

    def forward(self, x):
        x = F.relu(F.max_pool2d(self.conv1(x), 2))
        x = F.relu(F.max_pool2d(self.conv2_drop(self.conv2(x)), 2))
        x = x.view(-1, 320)
        x = F.relu(self.fc1(x))
        x = F.dropout(x, training=self.training)
        x = self.fc2(x)
        return x
```

```python
train_losses = []
train_counter = []
test_losses = []

model = ConvEncoder(10)

opt = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(5):
    model.train()
    
    for batch, labels in train_loader:
        opt.zero_grad()
        output = model(batch.to(device))
        
        loss = criterion(output, labels.to(device))
        loss.backward()
        opt.step()
        
    train_losses.append(loss.item())

    test(model)
```

```python
data = torch.utils.data.DataLoader(
        torchvision.datasets.MNIST('./data',
               transform=torchvision.transforms.ToTensor(),
               download=True),
        batch_size=128,
        shuffle=True)
```

```python
class Encoder(nn.Module):
    def __init__(self, latent_dim):
        super(Encoder, self).__init__()
        self.linear1 = nn.Linear(784, 512)
        self.linear2 = nn.Linear(512, 256)
        self.linear3 = nn.Linear(256, 128)
        self.linear4 = nn.Linear(128, 64)
        self.linear5 = nn.Linear(64, latent_dim)
        self.batch = nn.BatchNorm1d(128)
        self.act = nn.LeakyReLU(0.1)

    def forward(self, x):
        x = torch.flatten(x, start_dim=1)
        x = self.act(self.linear1(x))
        x = self.act(self.linear2(x))
        x = self.act(self.linear3(x))
        x = self.batch(x)
        return self.act(self.linear5(self.linear4(x)))
```

```python
class Decoder(nn.Module):
    def __init__(self, latent_dim):
        super(Decoder, self).__init__()
        self.linear1 = nn.Linear(latent_dim, 64)
        self.linear2 = nn.Linear(64, 128)
        self.linear3 = nn.Linear(128, 256)
        self.linear4 = nn.Linear(256, 512)
        self.linear5 = nn.Linear(512, 784)
        self.batch = nn.BatchNorm1d(256)
        self.act = nn.LeakyReLU(0.1)

    def forward(self, z):
        z = self.act(self.linear1(z))
        z = self.act(self.linear2(z))
        z = self.act(self.linear3(z))
        z = self.batch(z)
        z = self.act(self.linear4(z))
        z = torch.sigmoid(self.linear5(z))
        return z.reshape((-1, 1, 28, 28))
```

```python
# Classic autoencoder:
class Autoencoder(nn.Module):
    def __init__(self, latent_dims):
        super(Autoencoder, self).__init__()
        self.encoder = Encoder(latent_dims)
        self.decoder = Decoder(latent_dims)

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)
```

```python
class VAE(nn.Module):
    def __init__(self, encoder, decoder, latent_dim, device=device):
        super(VAE, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device
        self.mu = None
        self.std = None
        
    def sample_z(self, mu, logvar):
        epsilon = torch.randn(logvar.size())
        epsilon = Variable(epsilon, requires_grad=False).type(torch.FloatTensor).to(self.device)
        z = mu + logvar * epsilon
        return z
    
    def forward(self, x):
        self.mu, logvar = self.encoder(x)
        self.std = torch.exp(0.5 * logvar)
        z = self.sample_z(self.mu.to(self.device), self.std.to(self.device))
        decoder_out = self.decoder(z)
        return decoder_out, self.mu, logvar
```

```python
epochs = 20
autoencoder = Autoencoder(2).to(device)
opt = torch.optim.Adam(autoencoder.parameters(), lr=1e-4)
mse = nn.MSELoss()
autoencoder.train()

for epoch in range(epochs):
    for batch, label in data:
        batch = batch.to(device)
        opt.zero_grad()
        reconstruction = autoencoder(batch)
        loss = ((batch - reconstruction) ** 2).mean()
        #loss = mse(batch, reconstruction) #+ autoencoder.encoder.kl / (28 * 28) #((batch - reconstruction)**2).sum() + autoencoder.encoder.kl
        loss.backward()
        opt.step()
    print(f'Loss {loss} after {epoch} epoch')
```

```python
def plot_latent(autoencoder, data, num_batches=100):
    plt.figure(figsize=(12, 8))
    with torch.no_grad():
        for i, (x, y) in enumerate(data):
            z = autoencoder.encoder(x.to(device))
            z = z.to('cpu').detach().numpy()
            plt.scatter(z[:, 0], z[:, 1], c=y, cmap='tab10')

            if i > num_batches:
                plt.colorbar()
                break
```

```python
plot_latent(autoencoder, data)
```

```python
def plot_reconstructed(autoencoder, r0=(-0.5, 2.5), r1=(-0.5, 2.5), n=12):
    w = 28
    img = np.zeros((n*w, n*w))
    plt.figure(figsize=(10, 15))
    with torch.no_grad():
        for i, y in enumerate(np.linspace(*r1, n)):
            for j, x in enumerate(np.linspace(*r0, n)):
                z = torch.Tensor([[x, y], [0, 0]]).to(device)
                x_hat = autoencoder.decoder(z)[0]
                x_hat = x_hat.reshape(28, 28).to('cpu').detach().numpy()
                img[(n - 1 - i) * w:(n - 1 - i + 1) * w, j * w:(j + 1) * w] = x_hat
        plt.imshow(img, extent=[*r0, *r1])
```

```python
class VAEncoder(nn.Module):
    def __init__(self, latent_dim):
        super(VAEncoder, self).__init__()
        self.linear1 = nn.Linear(784, 512)
        self.linear2 = nn.Linear(512, 256)
        self.linear3 = nn.Linear(256, 128)
        self.linear4 = nn.Linear(128, 64)
        self.batch = nn.BatchNorm1d(128)
        self.act = nn.LeakyReLU(0.1)
        self.mu = nn.Linear(64, latent_dim)
        self.logvar = nn.Linear(64, latent_dim)

        self.N = torch.distributions.Normal(0, 1)
        self.N.loc = self.N.loc.to(device)
        self.N.scale = self.N.scale.to(device)
        self.kl = 0

    def forward(self, x):
        x = torch.flatten(x, start_dim=1)
        x = torch.flatten(x, start_dim=1)
        x = self.act(self.linear1(x))
        x = self.act(self.linear2(x))
        x = self.act(self.linear3(x))
        x = self.batch(x)
        x = self.act(self.linear4(x))
        mu =  self.mu(x)
        log_var = self.logvar(x)
        sigma = torch.exp(0.5 * log_var)
        z = mu + sigma * self.N.sample(mu.shape)
        #self.kl = (sigma ** 2 + mu ** 2 - torch.log(sigma) - 1/2).sum()
        #self.kl = torch.sum(sigma ** 2 + mu ** 2 - 2 * torch.log(sigma) - 1)
        self.kl = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
        
        #kl_loss = - torch.sum(1 + logvar - mu.pow(2) - logvar.exp(), -1).mean()
        return z
```

```python
# Classis autoencoder:
class VAE(nn.Module):
    def __init__(self, latent_dims):
        super(VAE, self).__init__()
        self.encoder = VAEncoder(latent_dims)
        self.decoder = Decoder(latent_dims)

    def forward(self, x):
        z = self.encoder(x)
        return self.decoder(z)
```

```python
epochs = 25
vae = VAE(2).to(device)
opt = torch.optim.RMSprop(vae.parameters(), lr=1e-4)
mse = nn.MSELoss()
vae.train()
losses = []

for epoch in range(epochs):
    for batch, label in data:
        batch = batch.to(device)
        opt.zero_grad()
        reconstruction = vae(batch)
        rec_loss = mse(batch, reconstruction)
        kl_loss = vae.encoder.kl / (28 * 28 * batch.shape[0])
        loss = rec_loss + kl_loss
        loss.backward()
        opt.step()
    losses.append(loss.item())
    #print(f'Loss {loss} after {epoch} epoch')
    plt.plot(losses[-100:])
    plt.title('{}/{} Reconstruction loss: {:.4f}, KL loss: {:.4f}'.format(epoch + 1, epochs, rec_loss, kl_loss))
    plt.show()
    clear_output(wait=True)
```

```python
from PIL import Image

def interpolate_gif(autoencoder, filename, x_1, x_2, n=100):
    autoencoder.eval()
    
    with torch.no_grad():
        z_1 = autoencoder.encoder(x_1)
        z_2 = autoencoder.encoder(x_2)

        z = torch.stack([z_1 + (z_2 - z_1) * t for t in np.linspace(0, 1, n)]).squeeze()

        interpolate_list = autoencoder.decoder(z)
        interpolate_list = interpolate_list.to('cpu').detach().numpy()*255

    images_list = [Image.fromarray(img.reshape(28, 28)).resize((256, 256)) for img in interpolate_list]
    images_list = images_list + images_list[::-1] # loop back beginning

    images_list[0].save(
        f'{filename}.gif',
        save_all=True,
        append_images=images_list[1:],
        loop=1)
```
